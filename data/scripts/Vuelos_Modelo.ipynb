```python
# Convertimos la columna a formato datetime si no lo está
df_vuelos['Fecha_Hora_Local_despegue'] = pd.to_datetime(df_vuelos['Fecha_Hora_Local_despegue'], errors='coerce')

# Obtenemos el nombre del día de la semana
df_vuelos['dia_nombre'] = df_vuelos['Fecha_Hora_Local_despegue'].dt.day_name()
```

```python
df_vuelos.head(2)
```

```python
# Obtenemos el número del día
df_vuelos['dia_numero'] = df_vuelos['Fecha_Hora_Local_despegue'].dt.day
```

```python
# Obtenemos el número del mes
df_vuelos['mes_numero'] = df_vuelos['Fecha_Hora_Local_despegue'].dt.month
```

```python
# Definimos la función para clasificar el periodo del día
# mañana (entre 5:00 y 11:59), tarde (entre 12:00 y 18:59) y noche (entre 19:00 y 4:59), en base a Fecha
def check_condition_2(hms):
    hora = hms.time()  # Extraemos solo la hora de la fecha completa

    if pd.to_datetime('05:00').time() <= hora <= pd.to_datetime('11:59').time():
        return 'mañana'
    elif pd.to_datetime('12:00').time() <= hora <= pd.to_datetime('18:59').time():
        return 'tarde'
    else:
        return 'noche'

# Aplicamos la función a la columna 'Fecha_Hora_Local_despegue'
df_vuelos['periodo_dia'] = df_vuelos['Fecha_Hora_Local_despegue'].apply(check_condition_2)
```

```python
# Convertimos la columna 'Tiempo_Vuelo' a formato timedelta
df_vuelos['Tiempo_Vuelo'] = pd.to_timedelta(df_vuelos['Tiempo_Vuelo'])

# Luego, obtenemos los minutos
df_vuelos['minutos_vuelo'] = df_vuelos['Tiempo_Vuelo'].dt.total_seconds() / 60
```

```python
from catboost import CatBoostRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
pd.options.display.float_format = '{:.4f}'.format
```

```python
# Eliminamos dos atributos no necesarios para este modelo y la variable dependiente
X1 = df_raw.drop(['Aeronave','Pasajeros','minutos_vuelo'], axis=1)
```

```python
y1 = df_raw.loc[:, df_raw.columns == "minutos_vuelo" ]
```

```python
# Dividir el conjunto de datos en entrenamiento y prueba
X_train_1, X_test_1, y_train_1, y_test_1 = train_test_split(X1, y1, test_size=0.2, random_state=42)
```

```python
# Especificar las columnas categóricas
categorical_features = ['Origen',	'Destino',	'Aerolinea',	'periodo_dia',	'dia_nombre',	'dia_numero',	'mes_numero']
```

```python
# Crear y entrenar el modelo
model_1 = CatBoostRegressor(cat_features=categorical_features, verbose=0)
model_1.fit(X_train_1, y_train_1)
```

```python
# Hacer predicciones
y_pred_test_1 = model_1.predict(X_test_1)
```

```python
# Evaluar el modelo
mse_1 = mean_squared_error(y_test_1, y_pred_test_1)
rmse_1  = pd.to_numeric(mse_1**(1/2), downcast='float')
r2_1 = r2_score(y_test_1 ,y_pred_test_1)
print(f'Mean Squared Error: {round(mse_1)}')
print(f'Root Mean Squared Error: {round(rmse_1)} minutos')
print(f"R2: ", round(r2_1,4))
```
