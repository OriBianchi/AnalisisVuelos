```python
# Importamos librerias
import pandas as pd
import numpy as np
from datetime import datetime
from datetime import timedelta
```

```python
# Configurar pandas para mostrar todas las columnas
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
# Cargar a dataframe archivos CSV
df_vuelos = pd.read_csv("/content/vuelos_eda.csv", low_memory=False, sep=',')
df_aeropuertos = pd.read_csv("/content/aeropuertos_eda.csv", sep=',')
```

```python
df_vuelos.shape
```

## Carga - Armado de Tablones

```python
if 'Año' in df_vuelos.columns:
  df_vuelos = df_vuelos.drop('Año', axis=1)

if 'PAX' in df_vuelos.columns:
  df_vuelos = df_vuelos.drop('PAX', axis=1)
```

```python
# Seleccionamos todos los despegues
df_despegues = df_vuelos[df_vuelos['Tipo Movimiento'] == 'Despegue']
df_despegues = df_despegues.rename(columns={'Tipo Movimiento':'Tipo_Movimiento',
                                            'Aeropuerto':'Origen',
                                            'Origen / Destino':'Destino',
                                            'Aerolinea Nombre':'Aerolinea'})
# Reordenar columnas
orden = ['Fecha','Fecha_Hora_Local', 'Tipo_Movimiento', 'Origen', 'Destino', 'Aerolinea', 'Aeronave', 'Pasajeros']
df_despegues = df_despegues[orden]
df_despegues.head()
```

```python
# Incorporamos datos de aeropuertos con relacion a la columna Origen
df_merged = pd.merge(df_despegues, df_aeropuertos[['local', 'denominacion','provincia']], left_on='Origen', right_on='local', how='left')
df_merged = df_merged.rename(columns={'denominacion': 'Ciudad_O',
                                      'provincia':'Provincia_O'})
df_merged.drop('local', axis=1, inplace=True)
df_merged.head()
```

```python
df_merged.shape
```

```python
# Incorporamos datos de aeropuertos con relacion a la columna Destino
df_merged = pd.merge(df_merged, df_aeropuertos[['local', 'denominacion','provincia']], left_on='Destino', right_on='local', how='left')
df_merged = df_merged.rename(columns={'denominacion': 'Ciudad_D',
                                      'provincia':'Provincia_D'})
df_merged.drop('local', axis=1, inplace=True)
df_merged.head()
```

```python
# Crear identificador de vuelo basado en atributos para luego merge con aterrizajes
filtro_despegue = df_merged['Tipo_Movimiento'] == 'Despegue'
df_merged.loc[filtro_despegue, 'Vuelo_ID'] = (
                df_merged.loc[filtro_despegue, 'Fecha'].astype(str) + '_' +
                df_merged.loc[filtro_despegue, 'Origen'] + '_' +
                df_merged.loc[filtro_despegue, 'Destino'] + '_' +
                df_merged.loc[filtro_despegue, 'Aerolinea'] + '_' +
                df_merged.loc[filtro_despegue, 'Aeronave'] + '_' +
                df_merged.loc[filtro_despegue, 'Pasajeros'].astype(str)
)
```

```python
df_merged.head()
```

```python
# Mostrar 1 registro en el cual se pueda ver un valor entero de la columna Vuelo_Id
# Assuming 'filtro_despegue' is already defined as in your code
example_record = df_merged.loc[filtro_despegue, 'Vuelo_ID'].iloc[0]
print(example_record)
```

#### Aterrizajes

```python
# Seleccionamos todos los aterrizajes
df_aterrizajes = df_vuelos[df_vuelos['Tipo Movimiento'] == 'Aterrizaje']
df_aterrizajes.head()
```

```python
# Crear identificador de vuelo basado en atributos para luego merge con aterrizajes
filtro = df_aterrizajes['Tipo Movimiento'] == 'Aterrizaje'
df_aterrizajes_copy = df_aterrizajes.copy()
df_aterrizajes_copy['Vuelo_ID'] = (
    df_aterrizajes_copy['Fecha'].astype(str) + '_' +
    df_aterrizajes_copy['Origen / Destino'] + '_' +
    df_aterrizajes_copy['Aeropuerto'] + '_' +
    df_aterrizajes_copy['Aerolinea Nombre'] + '_' +
    df_aterrizajes_copy['Aeronave'] + '_' +
    df_aterrizajes_copy['Pasajeros'].astype(str)
)
```

```python
df_aterrizajes = df_aterrizajes_copy.copy()
df_aterrizajes.head()
```

```python
# Unir df_merged con df_aterrizajes por la columna Vuelo_ID, solo trayendo 'Fecha_Hora_Local' de df_aterrizajes
df_unido = pd.merge(df_merged, df_aterrizajes[['Vuelo_ID', 'Fecha_Hora_Local']], on='Vuelo_ID', how='left', suffixes=('_despegue', '_aterrizaje'))
```

```python
df_unido.head()
```

```python
# Verificar si existen valores que no coincidieron
nan_count = df_unido['Fecha_Hora_Local_aterrizaje'].isnull().sum()
print(f"La columna 'Fecha_Hora_Local' tiene {nan_count} valores NaN.")
```

```python
# eliminar de df todos los nan de la columna 'Fecha_Hora_Local_aterrizaje'
df_preliminar_1 = df_unido.dropna(subset=['Fecha_Hora_Local_aterrizaje'])
```

```python
# obtener la diferencia horaria en el df_completo entre los campos 'Fecha_Hora_Local_aterrizaje' y 'Fecha_Hora_Local_despegue'
df_preliminar_2 = df_preliminar_1.copy()

# Convertir las columnas de fecha y hora a objetos datetime
df_preliminar_2['Fecha_Hora_Local_aterrizaje'] = pd.to_datetime(df_preliminar_1['Fecha_Hora_Local_aterrizaje'])
df_preliminar_2['Fecha_Hora_Local_despegue'] = pd.to_datetime(df_preliminar_1['Fecha_Hora_Local_despegue'])

# Calcular la diferencia de tiempo

df_preliminar_2['Tiempo_Vuelo'] = (df_preliminar_2['Fecha_Hora_Local_aterrizaje'] - df_preliminar_2['Fecha_Hora_Local_despegue']).apply(lambda x: str(x).split()[-1][-8:])

# Mostrar el DataFrame con la nueva columna
df_preliminar_2[['Fecha_Hora_Local_despegue','Fecha_Hora_Local_aterrizaje', 'Tiempo_Vuelo']].head()
```

```python
# Verificar inconsistencias de duracion de vuelo
df_preliminar_2[ pd.to_datetime(df_preliminar_2['Tiempo_Vuelo']) > pd.to_datetime('04:30:00')].count()
```

```python
# Eliminar inconsistencias
df_preliminar_3 = df_preliminar_2.copy()
df_preliminar_4 = df_preliminar_3[ (pd.to_datetime(df_preliminar_3['Tiempo_Vuelo']) < pd.to_datetime('04:30:00')) ]
df_preliminar_4.shape
```

### 1. Configuración Inicial e Importación de Librerías

Al comienzo, el notebook importa las librerías `pandas` y `numpy` para el manejo de datos, y `datetime` de `datetime` para trabajar con fechas y horas. También se configuran las opciones de visualización de `pandas` para mostrar todas las filas y columnas de los DataFrames, lo que es útil para la exploración de datos.

### 2. Carga de Datos

Se cargan dos archivos CSV, `vuelos_eda.csv` y `aeropuertos_eda.csv`, en dos DataFrames de pandas llamados `df_vuelos` y `df_aeropuertos` respectivamente. `df_vuelos` contiene información detallada de los vuelos, mientras que `df_aeropuertos` contiene datos de los aeropuertos.

### 3. Limpieza y Preparación Inicial de `df_vuelos`

Se eliminan las columnas 'Año' y 'PAX' de `df_vuelos`, si existen, para simplificar el DataFrame y enfocarse en los datos relevantes para el análisis posterior.

### 4. Procesamiento de Despegues

Se filtra `df_vuelos` para crear un nuevo DataFrame `df_despegues` que solo contiene los registros de tipo 'Despegue'. Luego, se renombran algunas columnas (`Tipo Movimiento`, `Aeropuerto`, `Origen / Destino`, `Aerolinea Nombre`) para que sean más descriptivas y coherentes en el contexto de un despegue (`Tipo_Movimiento`, `Origen`, `Destino`, `Aerolinea`). Finalmente, se reordenan las columnas para tener una estructura lógica.

### 5. Incorporación de Datos de Aeropuertos (Origen)

El DataFrame `df_despegues` se fusiona con `df_aeropuertos` usando la columna 'Origen' (código IATA del aeropuerto) para añadir información detallada del aeropuerto de origen, como la ciudad (`Ciudad_O`) y la provincia (`Provincia_O`). La columna auxiliar 'local' se elimina después de la fusión.

### 6. Incorporación de Datos de Aeropuertos (Destino)

De manera similar al paso anterior, el DataFrame resultante se vuelve a fusionar con `df_aeropuertos`, pero esta vez utilizando la columna 'Destino' para agregar la ciudad (`Ciudad_D`) y la provincia (`Provincia_D`) del aeropuerto de destino. La columna auxiliar 'local' también se elimina.

### 7. Creación de un Identificador Único de Vuelo (Vuelo_ID) para Despegues

Se genera una columna `Vuelo_ID` para cada registro de despegue. Este ID único se construye concatenando varios atributos del vuelo (Fecha, Origen, Destino, Aerolinea, Aeronave, Pasajeros). Esto es crucial para poder emparejar despegues con sus aterrizajes correspondientes más adelante.

### 8. Procesamiento de Aterrizajes

Se filtra el DataFrame original `df_vuelos` para crear `df_aterrizajes`, conteniendo solo los registros de tipo 'Aterrizaje'. Luego, al igual que con los despegues, se crea una columna `Vuelo_ID` con una lógica similar (Fecha, Origen / Destino, Aeropuerto, Aerolinea Nombre, Aeronave, Pasajeros) para poder unir los registros de aterrizaje con los de despegue.

### 9. Unificación de Despegues y Aterrizajes

Se realiza una fusión (`pd.merge`) entre el DataFrame de despegues enriquecido (`df_merged`) y el DataFrame de aterrizajes (`df_aterrizajes`) utilizando la columna `Vuelo_ID`. El objetivo principal es traer la `Fecha_Hora_Local` del aterrizaje al DataFrame de despegues, creando la columna `Fecha_Hora_Local_aterrizaje`. Se utilizan sufijos para diferenciar las columnas de fecha y hora locales de despegue y aterrizaje.

### 10. Manejo de Valores Faltantes y Limpieza Adicional

Se verifica cuántos registros de despegue no encontraron un aterrizaje correspondiente (es decir, tienen valores `NaN` en `Fecha_Hora_Local_aterrizaje`). Estos registros se eliminan para asegurar que solo se consideren vuelos completos, creando `df_preliminar_1`.

### 11. Cálculo del Tiempo de Vuelo

Las columnas de fecha y hora de despegue y aterrizaje se convierten a formato `datetime`. Luego, se calcula la diferencia entre estas dos columnas para obtener la duración del vuelo (`Tiempo_Vuelo`), que se formatea para mostrar solo la hora, minuto y segundo.

### 12. Filtrado de Inconsistencias en el Tiempo de Vuelo

Se detectan y eliminan vuelos con duraciones irrealmente largas (más de 4 horas y 30 minutos). Esto ayuda a limpiar el conjunto de datos de posibles errores de registro o emparejamiento. El DataFrame resultante, `df_preliminar_4`, contiene solo vuelos con duraciones consistentes.

### 13. Guardado del DataFrame Final

Finalmente, el DataFrame `df_preliminar_4`, que contiene los vuelos procesados con sus tiempos de vuelo calculados y las inconsistencias eliminadas, se guarda en un nuevo archivo CSV llamado `vuelos_preliminar.csv`.
